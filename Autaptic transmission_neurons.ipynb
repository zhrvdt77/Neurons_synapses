```python
#----------------------  (*neuron autapse*)  --------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
import math

#####################

data=[]


def timedoc(n):
    if n==M:
        return 1000000
    else:
        return np.random.exponential(1/(k*(M-n)))


def timedocd(nd):
    if nd==Md:
        return 1000000
    else:
        return np.random.exponential(1/(kd*(Md-nd)))
    


def timerelease():
    ct1=np.random.exponential(1/f)
    return ct1


def released(nd):
    return np.random.binomial(nd,p_d)

def release(n):
    return np.random.binomial(n,pr)


def voltage(v,ti):
    return(v*np.exp(-ti/tv))


    
    
vth=.05
v0=0
v=0
tv=.05;

f=100

M=20
k=10000
pr=.3
n=10
b=release(n)
cu=0.005


Md=20
kd=10000
p_d=1
nd=10
bd=released(nd)
cd=0.006
   
dt=.01
t=0




timer1=dt

timer2=timedoc(n)
timer3=timerelease()
timer4=timedocd(nd)


timer6counter=0



react=0
data=[[react,n,nd,t,b,bd,v,np.nan]]
tmax=1000
ts=[0]


while t<tmax:
    tr=np.min([timer1,timer2,timer3,timer4])
    react=np.argmin([timer1,timer2,timer3,timer4])+1
    t+=tr
    if react==1:
        timer1=dt
        timer2=timedoc(n)
        timer3=timerelease()
        timer4=timedocd(nd)
        timer6counter=timer6counter+tr
        v=voltage(v,tr)
        data.append([react,n,nd,t,np.nan,np.nan,v,np.nan])
        
    elif react==2:
        timer1=timer1-tr
        n=n+1
        timer2=timedoc(n)
        timer3=timerelease()
        timer4=timedocd(nd)
        timer6counter=timer6counter+tr
        v=voltage(v,tr)
        data.append([react,n,nd,t,np.nan,np.nan,v,np.nan])
        
    elif react==3:
        timer1=timer1-tr
        b=release(n)
        n=n-b
        v=voltage(v,tr)
        v=v+cu*b
        timer2=timedoc(n)
        timer3=timerelease()
        timer4=timedocd(nd)
        timer6counter=timer6counter+tr
        if v>=vth:
            bd=released(nd)
            v=0-cd*bd
            nd=nd-bd
            data.append([react,n,nd,t,b,bd,v,timer6counter])
            
            timer6counter=0
        else:
            data.append([react,n,nd,t,b,np.nan,v,np.nan])
            
        
        
    elif react==4:
        timer1=timer1-tr
        timer2=timedoc(n)
        timer3=timerelease()
        nd=nd+1
        timer4=timedocd(nd)
        v=voltage(v,tr)
        timer6counter=timer6counter+tr
        data.append([react,n,nd,t,np.nan,np.nan,v,np.nan])
               

    
df=pd.DataFrame(data,columns=['React','n','nd','Time','b','bd','v','fpt'])
```
